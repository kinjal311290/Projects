```python
# Plot Total Sales Per Month for the Year 2011.  How the total sales have increased over months in the Year 2011.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month

df_2011 = df[df['Year'] == 2011]

monthly_sales = df_2011.groupby('Month')['Amount'].sum()

# Plot
plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='b')
plt.title('Total Sales Per Month in 2011')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.xticks(range(1,13))
plt.grid(True)
plt.show()

# Find and print month with lowest sales
lowest_month = monthly_sales.idxmin()
lowest_sales = monthly_sales.min()
print(f"Month with lowest sales: {lowest_month} (Total Sales: {lowest_sales:.2f})")
```

```python
import matplotlib.pyplot as plt

# If not already calculated, create 'InvoiceAmount' column if needed:
# Here, I assume 'Amount' is the invoice amount per row, but if you want total invoice amount per InvoiceNo, then:

invoice_amounts = df.groupby('InvoiceNo')['Amount'].sum().reset_index()

plt.figure(figsize=(10,6))

plt.scatter(
    invoice_amounts.index,
    invoice_amounts['Amount'],
    c='purple',  # change color of scatter points
    alpha=0.6,
    edgecolor='w',
    s=50
)

plt.title('Scatter Plot of Invoice Amounts')
plt.xlabel('Invoice Index')
plt.ylabel('Invoice Amount')
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
