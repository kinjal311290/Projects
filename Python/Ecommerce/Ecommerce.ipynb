```python
# Name field contains full name –use a regular expression to separate title, first name, last name
```

```python
import re

def split_name(full_name):
    pattern = r'^(Mr\.|Mrs\.|Ms\.|Dr\.|Miss|Sir|Mx\.)\s+(\w+)\s+(\w+)$'
    
    match = re.match(pattern, full_name.strip())
    if match:
        title = match.group(1)
        first_name = match.group(2)
        last_name = match.group(3)
        return title, first_name, last_name
    else:
        return None, None, None


names = [
    "Mr. Owen Harris",
    "Dr. Jane Doe",
    "Mrs. Emily Smith",
    "Ms. Anna Lee",
    "Sir John Doe"
]

for name in names:
    title, first, last = split_name(name)
    print(f"Full Name: {name} => Title: {title}, First Name: {first}, Last Name: {last}")
```

```python
# Pass a customer to function "createOrder" and throw CustomerNotAllowedException in case of blacklisted value is 1
```

```python
class CustomerNotAllowedException(Exception):
    def __init__(self, message="Customer is not allowed"):
        self.message = message
        super().__init__(self.message)

class Customer:
    def __init__(self, title, first_name, last_name, blacklisted=0):
        self.title = title
        self.first_name = first_name
        self.last_name = last_name
        self.blacklisted = blacklisted

def createOrder(customer, product_name, product_code):
    if customer.blacklisted == 1:
        raise CustomerNotAllowedException(f"Customer {customer.first_name} {customer.last_name} is blacklisted and not allowed to place orders.")

    print(f"Order created for {customer.first_name} {customer.last_name}: Product - {product_name} (Code: {product_code})")


try:
    cust = Customer("Mr.", "Owen", "Harris", blacklisted=1)
    createOrder(cust, "Laptop", "LP1001")
except CustomerNotAllowedException as e:
    print(e)

try:
    cust2 = Customer("Ms.", "Anna", "Lee", blacklisted=0)
    createOrder(cust2, "Smartphone", "SP2002")
except CustomerNotAllowedException as e:
    print(e)
```

```python
# Create Class OrderReturn object of type Order in case customer is eligible
```

```python
class CustomerNotAllowedException(Exception):
    def __init__(self, message="Customer is not allowed"):
        self.message = message
        super().__init__(self.message)

class Customer:
    def __init__(self, title, first_name, last_name, blacklisted=0):
        self.title = title
        self.first_name = first_name
        self.last_name = last_name
        self.blacklisted = blacklisted

class Order:
    def __init__(self, customer, product_name, product_code):
        if customer.blacklisted == 1:
            raise CustomerNotAllowedException(f"Customer {customer.first_name} {customer.last_name} is blacklisted and cannot place orders.")
        self.customer = customer
        self.product_name = product_name
        self.product_code = product_code
    
    def __str__(self):
        return f"Order: {self.product_name} ({self.product_code}) for {self.customer.first_name} {self.customer.last_name}"

class OrderReturn(Order):
    def __init__(self, customer, product_name, product_code, reason):

        super().__init__(customer, product_name, product_code)
        self.reason = reason
    
    def __str__(self):
        return f"OrderReturn: {self.product_name} ({self.product_code}) for {self.customer.first_name} {self.customer.last_name}, Reason: {self.reason}"


try:
    cust = Customer("Ms.", "Anna", "Lee", blacklisted=0)
    order_return = OrderReturn(cust, "Smartphone", "SP2002", "Defective product")
    print(order_return)
except CustomerNotAllowedException as e:
    print(e)

try:
    cust2 = Customer("Mr.", "Owen", "Harris", blacklisted=1)
    order_return2 = OrderReturn(cust2, "Laptop", "LP1001", "Changed mind")
    print(order_return2)
except CustomerNotAllowedException as e:
    print(e)
```
